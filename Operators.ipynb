```python
char_input = input("Enter a single character: ")

if len(char_input) == 1:
    if char_input.isdigit():
        print(f"'{char_input}' is a digit.")
    elif char_input.islower():
        print(f"'{char_input}' is a lowercase alphabet character.")
    elif char_input.isupper():
        print(f"'{char_input}' is an uppercase alphabet character.")
    else:
        print(f"'{char_input}' is neither a digit nor an alphabet character.")
else:
    print("Please enter only a single character.")
```

```python
n = int(input("Enter any number : "))

# Using a list comprehension to generate the multiplication table strings
multiplication_table = [f" {n} x {i} = {n*i}" for i in range(1, 11)]

# Print each item from the generated list
for item in multiplication_table:
    print(item)
```

```python
import math

# Get a number from the user
number = int(input("Enter a non-negative integer to find its factorial: "))

if number < 0:
    print("Factorial is not defined for negative numbers.")
elif number == 0:
    print("The factorial of 0 is 1.")
else:
    # Calculate factorial using math.factorial()
    result = math.factorial(number)
    print(f"The factorial of {number} is: {result}")
```

```python
class Car:
  def __init__(self, model, color, company, year):
    self.color = color
    self.model = model
    self.company = company

 def display_info(self):
   print(f"Car: {self.company} {self.model} {self.color} ({self.year}"))

car1 = Car("Toyota", "Camry", "Red", 2022)
car2 = Car("Honda", "Civic", "Blue", 2021)

car1.display_info()
car2.display_info()
```

```python
class Student:
  def __init__(self, name, age, grade):
    self.name = name
    self.age = age
    self.grade = grade
  def display_info(self):
    print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")
  student1 = Student("Alice", 16, "A")
  student2 = Student("Bob", 17, "B")

  student1.display_info()
  student2.display_info()
```
