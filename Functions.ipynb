```python
x=10
def funct1():
  x=globals()['x']
  x=10
  x= x*10
  globals()['x']=x
  print(x)

funct1()
print(x)
```

```python
x=10
def funct1():
  global x
  x= x*100
  print(x)

funct1()
print(x)
```
